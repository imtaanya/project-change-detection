```python
import os
import zipfile
from snappy import ProductIO, GPF, HashMap

# --- USER INPUTS ---
data_path = "./Downloads/change-detection/downloads/S1A_IW_GRDH_1SDV_20250918T004352_20250918T004417_061035_079B11_FD56"
aoi_path = "aoi1.geojson"
output_name = "S1_clipped"

# --- AUTO-DETECTION ---

if data_path.lower().endswith(".zip"):
    print("üì¶ ZIP file detected. Extracting...")
    extract_dir = os.path.splitext(data_path)[0] + "_unzipped"
    os.makedirs(extract_dir, exist_ok=True)

    with zipfile.ZipFile(data_path, "r") as zip_ref:
        zip_ref.extractall(extract_dir)
    
    # Find the .SAFE folder inside the unzipped directory
    safe_dirs = [d for d in os.listdir(extract_dir) if d.endswith(".SAFE")]
    if not safe_dirs:
        raise FileNotFoundError("No .SAFE folder found inside ZIP extraction.")
    safe_path = os.path.join(extract_dir, safe_dirs[0])
    print(f"‚úÖ Unzipped to: {safe_path}")

elif data_path.lower().endswith(".safe") and os.path.isdir(data_path):
    print("üìÅ SAFE folder detected.")
    safe_path = data_path

else:
    raise FileNotFoundError("‚ùå Input must be either a .zip file or a .SAFE folder path.")

# --- PROCESSING WITH SNAP (snappy) ---

manifest_path = os.path.join(safe_path, "manifest.safe")

if not os.path.exists(manifest_path):
    raise FileNotFoundError(f"manifest.safe not found at: {manifest_path}")

print(f"üìÑ Reading Sentinel-1 product: {manifest_path}")
product = ProductIO.readProduct(manifest_path)

# --- CLIP USING AOI ---
parameters = HashMap()
parameters.put('geoRegion', aoi_path)

print(f"‚úÇÔ∏è Clipping product to AOI from: {aoi_path}")
subset = GPF.createProduct('Subset', parameters, product)

# --- WRITE OUTPUT ---
print("üíæ Writing clipped product...")
ProductIO.writeProduct(subset, output_name, 'GeoTIFF')

print(f"‚úÖ Done! Output saved as {output_name}.tif")
```
